```python
#begin table
print("<table>")

#column headers
print("<th>")
print("<td>ID</td>")
print("<td>Year</td>")
print("<td>Round</td>")
print("<td>Circuit</td>")
print("<td>Name</td>")
print("<td>Date</td>")
print("<td>Time</td>")
print("<td>Wiki</td>")
print("</th>")

infile = open("races.csv")

for line in infile:
    row = line.split(",")
    race = row[0]
    year = row[1]
    rnd = row[2]
    circuit = row[3]
    name = row[4]
    date = row[5]
    time = row[6]
    wiki = row[7]
    
    print("<tr>")
    print("<td>%s</td>" % race)
    print("<td>%s</td>" % year)
    print("<td>%s</td>" % rnd)
    print("<td>%s</td>" % circuit)
    print("<td>%s</td>" % name)
    print("<td>%s</td>" % date)
    print("<td>%s</td>" % time)
    print("<td>%s</td>" % wiki)
    print("</tr>")
    
#end table
print("</table>")
```
